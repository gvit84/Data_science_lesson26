Побудуємо дискретний статистичний розподіл вибірки:

```python
tabledata = [[6, 7, 8, 9, 10, 11, 12, 13, 14],
            [1, 3, 6, 8, 6, 6, 5, 3, 2],
            [0.025, 0.075, 0.15,  0.2, 0.15, 0.15, 0.125, 0.075, 0.05]]
```

```python
pd.DataFrame(tabledata, 
             index=["xi", "ni", "Wi"],
             columns=[1, 2, 3, 4, 5, 6, 7, 8, 9])
```

```python
xi_list = tabledata[0]
```

```python
ni_list = tabledata[1]
```

```python
Wi_list = tabledata[2]
```

```python
plt.title("Полігон частот")
plt.xlabel("$x_i$")
plt.ylabel("$n_i$")
plt.plot(xi_list, ni_list)
```

```python
plt.title("Полігон відносних частот")
plt.xlabel("$x_i$")
plt.ylabel("$W_i$")
plt.plot(xi_list, Wi_list)
```

```python
plt.show()
```

$F^*(X) = W(X<x) = n_x / n =$                              

                              0              x ≤ 6
                              0.025      6 < x ≤ 7
                              0.1        7 < x ≤ 8
                              0.25       8 < x ≤ 9
                              0.45       9 < x ≤ 10
                              0.6       10 < x ≤ 11
                              0.75      11 < x ≤ 12
                              0.875     12 < x ≤ 13

                              1              x > 14
